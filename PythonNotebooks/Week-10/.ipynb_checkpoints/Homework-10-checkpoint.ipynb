```python
import arcpy
```

```python
single_raster = 'rasters/CHIRP.1981.01.1.tif'
raster_info = arcpy.Describe(single_raster)
print(raster_info.format)
print(raster_info.spatialReference.name)
```

```python
rasterList = []

for i in range(12):
    for j in range(3):
        if (i < 9):
            rasterFile = 'rasters/CHIRP.1981.0' + str(i + 1) + '.' + str(j + 1) + '.tif'
        else: 
            rasterFile = 'rasters/CHIRP.1981.' + str(i + 1) + '.' + str(j + 1) + '.tif'
        rasterArray = arcpy.RasterToNumPyArray(rasterFile)
        rasterList.append(rasterArray)
        
print(rasterList)        
```

```python
import numpy as np
```

```python
concat = np.stack(rasterList)
```
