## Question 1:

```python
import pandas as pd
```

```python
quake_csv = pd.read_csv('all_day.csv')
```

```python
df = pd.DataFrame(quake_csv)
```

```python
new_df = df.drop(columns=['nst', 'gap', 'dmin', 'rms', 'net'])
```

```python
locations = new_df.iloc[:,1:3].values.tolist()
```

```python
import arcgis
from arcgis.features import SpatialDataFrame
```

```python
sdf = SpatialDataFrame(new_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```

```python
new_df.append(sdf)
```

```python
sdf1 = sdf[sdf['mag'] < 2.5]
```

```python
sdf2 = sdf[(sdf['mag']>2.5) & (sdf['mag'] < 4)]
```

```python
sdf3 = sdf[(sdf['mag'] > 4) & (sdf['mag'] < 5)]
```

```python
sdf4 = sdf[sdf['mag'] > 5]
```

```python
sdf1.reset_index(inplace=True, drop=True)
```

```python
sdf2.reset_index(inplace=True, drop=True)
```

```python
sdf3.reset_index(inplace=True, drop=True)
```

```python
sdf4.reset_index(inplace=True, drop=True)
```

## Question 4

```python
year = 2015
month = 'January'
country = 'FRA'
```

```python
def month_string_to_number(string):
    m = {
        'jan': '1',
        'feb': '2',
        'mar': '3',
        'apr':'4',
         'may':'5',
         'jun':'6',
         'jul':'7',
         'aug':'8',
         'sep':'9',
         'oct':'10',
         'nov':'11',
         'dec':'12'
        }
    s = string.strip()[:3].lower()

    try:
        out = m[s]
        return out
    except:
        raise ValueError('Not a month')
```

```python
month_string_to_number(month)
```

```python
url = 'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=' + str(year) + '&month=' + month_string_to_number(month) + '&coa=' + country
data = requests.get(url)
as_json = data.json()
df3 = pd.DataFrame(as_json)
```

```python
df3.sort_values(by='value')
```
